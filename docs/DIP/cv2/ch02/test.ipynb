```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

fig, ax = plt.subplots(figsize=(6,5))
matplotlib.rcParams['pgf.texsystem'] = 'pdflatex'
matplotlib.rcParams.update({'font.family': 'serif', 'font.size': 18,
    'axes.labelsize': 20,'axes.titlesize': 24, 'figure.titlesize' : 28})
matplotlib.rcParams['text.usetex'] = True
r = np.linspace(1.,100.,1000)
f = r/((r+1)**2)
ax.plot(r,f)
ax.grid(color='k', linestyle='--', linewidth=1)
ax.set_xticks([1.0, 5.0, 9.0, 13.0, 17.0])
ax.set_xlim([1.0, 19])
ax.set_xlabel('$r$')
ax.set_ylabel(r'$f=\frac{r}{(1+r)^2}$')
```
